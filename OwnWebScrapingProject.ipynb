# Own WebScraping Project


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_average_wage"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
table= soup.find_all('table')[0]
```

```python
headers = table.find_all('th')
print(headers)
```

```python
country_titles = [finalHeader.text.strip() for finalHeader in headers]

print(country_titles)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns = country_titles)
df
```

```python
table_data= table.find_all('tr')
```

```python
print(table_data)
```

```python
for row in table_data[1:]:
    row_data = row.find_all('td')
    new_data = [data.text.strip() for data in row_data]
    cleaned_data = [item.replace('\u202f*', '') for item in new_data]
    print(cleaned_data)
    

 
```

```python
for row in table_data[1:]:
    row_data = row.find_all('td')
    new_data = [data.text.strip() for data in row_data]
    cleaned_data = [item.replace('\u202f*', '') for item in new_data]
    length = len(df)
    df.loc[length] = cleaned_data
    print(cleaned_data)
    
```
